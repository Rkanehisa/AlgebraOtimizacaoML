```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import TruncatedSVD
```

```python
data = np.load("X.npy")
```

```python
X = np.reshape(data[0],[50,37])
```

```python
plt.imshow(X,cmap="Greys")
```

```python
u2, s2, vh2 = np.linalg.svd(X, full_matrices=False)
d = np.diag(s2)
c = np.dot(np.dot(u2, d), vh2)
error = X - c
maxError = np.amax(abs(error))
```

```python
f,axarr = plt.subplots(1,3)
axarr[0].imshow(X,cmap="Greys")
axarr[1].imshow(c,cmap="Greys")
axarr[2].imshow(error,cmap="Greys")
```
